```python
# Set up query parameters
tiers = ['CHALLENGER', 'GRANDMASTER', 'MASTER', 'DIAMOND', 'PLATNIUM', 'GOLD', 'SILVER', 'BRONZE', 'IRON']
divisions = ['I', 'II', 'III', 'IV']
```

```python
# Now I can collect match data! First step is to create the empty dataframe

# Send a request
r = requests.get('https://na1.api.riotgames.com/lol/match/v4/matches/3239984916?api_key=RGAPI-69860d32-9db7-4d31-ad8b-1b352cd8e570')

# Create the dataframes (Participant and Bans)
part_matchdat_df=pd.json_normalize(data=r.json(), record_path='participants', meta=['gameId', 'seasonId', 'gameDuration'])
bans_matchdat_df=pd.json_normalize(data=r.json(), record_path=['teams', 'bans'])

# Rename bans_matchdat_df columns
bans_matchdat_df.columns = ['Banned Champion Id', 'pickTurn']

# Define a function for the pictTurn column value for the part_matchdat_df
def f(x):
    if x['participantId'] == 1: return 1
    elif x['participantId'] == 2: return 4
    elif x['participantId'] == 3: return 5
    elif x['participantId'] == 4: return 8
    elif x['participantId'] == 5: return 9
    elif x['participantId'] == 6: return 2
    elif x['participantId'] == 7: return 3
    elif x['participantId'] == 8: return 6
    elif x['participantId'] == 9: return 7
    elif x['participantId'] == 10: return 10
    
# Create pickTurn column for part_matchdat_df
part_matchdat_df['pickTurn'] = part_matchdat_df.apply(f, axis=1)

# Merge the two dataframes together on pickTurn
matches_data_df = pd.merge(part_matchdat_df, bans_matchdat_df, on= 'pickTurn')

# Remove all rows to create the empty dataframe 
matches_data_df = matches_data_df.iloc[0:0]

# Show empty dataframe
matches_data_df
```

```python
# Now we can request for match data using the GameIDs in gameID
# This will take about 3 hours and 15 min to run (Based on amount of games)
# Create a for loop to iterate over
# Lets do loops of 1500 due to gateway timeouts

counter=1

for i in range(0,1500):

    # Send a request
    r = requests.get('https://na1.api.riotgames.com/lol/match/v4/matches/' + game_ids_str[i] + '?api_key=' + api_key)
    print(r)
    print(counter)
    # Create the dataframes (Participant and Bans)
    part_matchdat_df=pd.json_normalize(data=r.json(), record_path='participants', meta=['gameId', 'seasonId', 'gameDuration'])
    bans_matchdat_df=pd.json_normalize(data=r.json(), record_path=['teams', 'bans'])

    # Rename bans_matchdat_df columns
    bans_matchdat_df.columns = ['Banned Champion Id', 'pickTurn']
    
    # Create pickTurn column for part_matchdat_df
    part_matchdat_df['pickTurn'] = part_matchdat_df.apply(f, axis=1)

    # Merge the two dataframes together on pickTurn
    Intermediate_df = pd.merge(part_matchdat_df, bans_matchdat_df, on= 'pickTurn')
    
    # Drop all rows not containing Nasus
    indexNames = Intermediate_df[Intermediate_df['championId'] != 75].index
    Intermediate_df.drop(indexNames, axis=0, inplace=True)
    
    # Append the list
    matches_data_df = matches_data_df.append(Intermediate_df, ignore_index = True)
    
    counter = counter+1
    time.sleep(0.87)  
```
